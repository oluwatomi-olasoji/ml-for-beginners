Task: Using BMI as a Predictor of Disease progression of diabetes 
1. Using linear regression 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

```python
# LOAD the diabetes dataset
X, y = datasets.load_diabetes(return_X_y= True) #X are the features, y is the label (target variable)

#printing the shape of the dataset
print(X.shape)
print(X[0])
X.dtype

```

```python
#y.head()
```

```python
#Extracting the BMI column which is at index 2
X = X[:,2]
print(X.shape)
```

```python
print(X)
```

```python
#reshaping x into a 2d array, to train the regression model
X = X.reshape(-1,1)
print(X.shape)
print(X)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(X,y, test_size= 0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
```

```python
#getting predictions
y_pred = model.predict(x_test)
```

```python
#Visualizin the results with a graph
plt.scatter(x_test, y_test, color = 'black')

plt.plot(x_test, y_pred, color = 'blue',linewidth = 3)

plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A graph plot showing dabetes Progression against BMI (predicted and actual)')

plt.show()
```
